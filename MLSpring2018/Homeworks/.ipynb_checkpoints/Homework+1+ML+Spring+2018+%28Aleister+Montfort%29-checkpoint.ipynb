```python
def create_df(variable):

    df = combined_df.loc[(combined_df['Type of Service Request'] == variable), ['creation_date', 'counter']]
    df['creation_date'] = pd.to_datetime(df['creation_date'])
    serie_graf = df.set_index('creation_date')['counter']
    collapsed = serie_graf.resample('A').count()
    collapsed= collapsed.loc['2010-12-31': '2017-12-31']
    collapsed.name = variable
    return collapsed
```

```python
varlist = ['Graffiti Removal', 'Alley Light Out', 'Vacant/Abandoned Building']
for element in varlist:
    df = create_df(element)
    plt.figure(); df.plot(style='k--'); plt.title('Annual 311 requests', fontsize=11); \
    plt.suptitle(df.name, fontsize=15)
    plt.show()
```

```python
def time_df(variable):
    creation = combined_df.loc[(combined_df['Type of Service Request'] == variable),['creation_date']]
    completion = combined_df.loc[(combined_df['Type of Service Request'] == variable),['completion_date']]
    creation_timess = pd.to_datetime(creation['creation_date'])
    completion_timess = pd.to_datetime(completion['completion_date'])
    difference =  (completion_timess - creation_timess)
    single_seriesdf = pd.concat([creation_timess, completion_timess, difference], axis = 1)
    single_seriesdf.columns = ['creation_date', 'completion_date', 'difference']
    single_seriesdf['creation_date'] = pd.to_datetime(single_seriesdf['creation_date'])
    serie_graf = single_seriesdf.set_index('creation_date')['difference']
    serie_graf= serie_graf.loc['2010-12-31': '2017-12-31']
    serie_graf = serie_graf.dt.days
    collapsed = serie_graf.resample('A').mean()
    collapsed.name = variable
    return collapsed


```

```python
varlist = ['Graffiti Removal', 'Alley Light Out', 'Vacant/Abandoned Building']
for element in varlist:
    df = time_df(element)
    plt.figure(); df.plot(style='k--'); plt.title('Average Annual Time Response to Requests (Days)', fontsize=11); \
    plt.suptitle(df.name, fontsize=15)
    plt.show()
```

```python
combined_df['year'] = pd.DatetimeIndex(combined_df.loc[:,'creation_date']).year
combined_df['month'] = pd.DatetimeIndex(combined_df.loc[:,'creation_date']).month
```

```python
def df_2017(variable):
    df = combined_df.loc[(combined_df["year"]==2017) & \
                         (combined_df["Type of Service Request"]== variable), ["ZIP Code","counter"]]
    df= df.groupby(['ZIP Code'])['counter'].count()
    df.name = variable
    return df
```

```python
def hist_by_ZIP(df):
    plt.figure();
    df.hist()
    plt.xlabel('Requests')
    plt.ylabel('Frequency')
    plt.title('Histogram of'+' '+ df.name +' '+ '2017')
    plt.grid(True)
    plt.show()
```

1. Among the three different types of requests, graffiti removal requests are the most common type of request in the City of Chicago. Since 2011, this type of request accounts for around 140,000 requests per year and the trend along these past 5 years remains pretty stable.
2. Alley light out requests are smaller in magnitude than graffiti removals. In 2011, it reached a peak (of around 40,000). Then it decreased to around 20,000 in 2013, and started to show an increasing trend to reach around 30,000 requests per year.
3. Vacant and abandoned building requests represent the smaller amount of requests, and it has showed a sistematic decrease since 2011. In this year, requests were more than 14,000, and in 2017, this numbern dropped to around 4,000, which is a reduction of aroun 30%.
4. Regarding the state capacity to attend these demands, then City of Chicago has done a good job in reducing the time response of graffiti removals request. The time to solve these demands dramatically dropped from 8 days in 2012, to only one day in 2017. In the case of Alley lights out, time response has showed a more erratic trend. It dropped from 60 days in 2011 to only 20 in 2013, but has increased to more than 30 in 2017. We don't have any information on time response for abandoned buildings.
5. When we analyze how requests are distributed among neigborhoods (that share the same ZIP Code), we can make some observations from the descriptive statistics shown above. All three ditributions are right skewed (median is to the left of the mean), suggesting a skewed distributution and that a few neighborhoods concentrate higher values of requests. In the case of graffiti removals, the median neighborhood makes 721 requests per year, whereas the average neighborhood has 1,848 requests. For vacant buildings, these numbers are 14.5 and 70. Alley lights out requests show a more even distribution, suggesting that the problem is more widespread among different neighborhoods. In this case, the median request is 336 and the mean is 448.
